Let's try it. We create two variables and assign two values to them. 

```python
a = 10
b = 20
```

**Addition**

```python
addition = a + b
print(addition)
```

**Subtraction**

```python
substraction = b - a
print(substraction)
```

**Multiplication**

```python
multiplication = a * b
print(multiplication)
```

**Division**  
Note that in this case, python returns a float.

```python
division = b / a
print(division)
```

**Floor division**  
To obtain an integer, you can use the floor division operator.

```python
floor_division = b // a
floor_division_bis = 31 // 10

print("floor_division: ", floor_division)
print("floor_division_bis: ", floor_division_bis)  # Outputs 3, not 3.1
```

**Modulus**  
To obtain the rest of a division.

```python
modulus = b % a
print("modulus :", modulus)
modulus_bis = 21 % 2
print("modulus_bis :", modulus_bis)
```

**Exponent**  
10 to the power 2

```python
power_of_two = a ** 2
print(power_of_two)
```

```python
print(a == b)
print(a == 10)
```

**Not equals**

```python
print(a != b)
print(a != 10)
```

```python
print(a > b)
print(b > a)
```

```python
print(a < b)
print(b < a)
```

```python
print(a <= b)
print(b <= a)
```

**Assignment**

```python
a = 10
name = "Alan Turing"
print(a)
print(name)
```

**Add AND** (`+=`)

```python
a += 10
name += " is a good mathematician"
print(a)
print(name)
```

**❗ But beware, we can only add variables of the same type**  
The variable `a` is of type integer and the variable `name` is of type string.

```python
name += a  # It's the same as `name = name + a`
```

```python
a += name  # It's the same as `a = a + name`
```

**Subtract AND** (`-=`)

```python
a = 20
a -= 10
print(a)
```

This operator does not work with strings, unlike the `+=` operator.

```python
name -= "Alan Turing"
```

**Multiply AND** (`*=`)  
This operator works with strings.

```python
a = 10
a *= 10
print(a)

text = "Alan Turing"
text *= 10
print(text)
```

**Divide AND** (`/=`)

```python
a = 100
a /= 10
print(a)
```

This operator does not work with strings, unlike the `*=` operator.  
Note that in this case, Python returns a float.

**Modulus AND** (`=`)

```python
a = 100
a %= 3
print(a)
```

This operator does not work with strings.

**Exponent AND** (`**=`)

```python
a = 2
a **= 3  # ( 2 * 2 * 2)
print(a)
```

This operator does not works with character strings.

**Floor Division AND** (`//=`)

```python
a = 20
a //= 3
print(a)
```

Note: This operator does not work with strings
